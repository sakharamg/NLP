```python
#Get the type of data structure
listType=type(semcor.tagged_sents(tag='sem')[0][0])
treeType=type(semcor.tagged_sents(tag='sem')[0][1])
print(type(semcor.tagged_sents(tag='sem')[0][0]))
print(type(semcor.tagged_sents(tag='sem')[0][1]))
synText=[]
synTag=[]
current_index=0
for taggedSents in semcor.tagged_sents(tag='sem'):
    synText.append([])
    synTag.append([])
    for phrase in taggedSents:
#         print(phrase)
        if type(phrase)==list:
            synText[current_index].append(phrase[0]) #if its a list fetch the word
            synTag[current_index].append('UNK')
        else:
            try:
                synTag[current_index].append(phrase.label().synset().offset())
            except:
                synTag[current_index].append('NOSYNSETID')
            for nodes in phrase:
                if type(nodes)==nltk.tree.Tree:
                    str=""
                    for node in nodes:
                        str+=" "+node
                    str.strip()
                    synText[current_index].append(str)
                else:
                    synText[current_index].append(nodes)
    current_index+=1
no_sents=current_index
```
